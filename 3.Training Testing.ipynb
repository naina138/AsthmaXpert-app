```python

from sklearn.model_selection import train_test_split
import pandas as pd


```

```python
# Step 2: Load cleaned dataset (if not already loaded)
df = pd.read_csv("asthma_cleaned_dataset.csv")

```

```python
# Step 3: Split into Features (X) and Target (y)
X = df.drop(columns=["Diagnosis"])  # All input features
y = df["Diagnosis"]                 # Target variable

```

```python
#  Step 4: Train-Test Split (80% train, 20% test, stratified for class balance)
X_train, X_test, y_train, y_test = train_test_split(
    X, y, test_size=0.2, stratify=y, random_state=42
)
```

```python

#  Step 5: Recombine X and y for saving
train_df = pd.concat([X_train, y_train], axis=1)
test_df = pd.concat([X_test, y_test], axis=1)
```

```python
# Save the training and testing sets
train_df.to_csv("asthma_train.csv", index=False)
test_df.to_csv("asthma_test.csv", index=False)

print("✅ Files saved:")
print("- asthma_train.csv")
print("- asthma_test.csv")
```
